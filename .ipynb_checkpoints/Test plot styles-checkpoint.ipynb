```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
x = np.linspace(-10,10,1000)
sinx = np.sin(x)
cosx = np.cos(x)
cos3x = np.cos(3*x)
sin3x = np.cos(3*x)

plt.plot(sinx,label='Sin(x)')
plt.plot(cosx,label='Cos(x)')
plt.plot(cos3x,label='Cos(3x)')
plt.plot(sin3x,label='Sin(x)')
plt.legend()
```
