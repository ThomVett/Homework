In order to map the university name to the canton, we need to use the Full search Geonames API (api.geonames.org/search). This API allows us to get xml data containing multiple location information related to the specific search term. By crawling in the xml code using the module etree, we are able to retrieve all these informations. In this information when the fcode is equal to "ADM1" it means that the entry is a canton (in switzerland specifically, See: http://www.geonames.org/export/codes.html).


In order to get the maximum number of succesful Canton retrieval we used different strategies. The first is to split each university name of the dataframe we previously created into each of its term. Then we make a geonames request for each of this term. In each entry retrieved, we look for the tag containing the fcode and see if its equal to "ADM1". If it is the case we crawl to the tag containing the name (which is the canton name) and return it.

This method effectively find the right canton for every university entry in the dataframe that contains the name of the canton (such as "Université de Berne") but will not retrieve the canton name of other entries (such as "AO Research Institute - AORI") or even entries that contains a major city (such as "EPFL école polytechnique de Lausanne").

To try to get the entries with city names in it, we performed a second strategy after the first. So every time that the first strategy doesn't return a canton name for an entry, we try the second strategy. We start the same, by spliting terms. Then we look for cities in retrieved entries. We gather the latitude and longitude of the first city we found and then we used another Geonames API, namely the findnearbyextended toponym, to find the canton. This API return the maximum information available about the coordonates we gave, among which there is an fcode ADM1. We then retrieve this canton name. This strategy succesfully retrieve entry containing the term "Lausanne" for example.



Definition of a functions that will be used for each strategies. 

query_term query the full text search the geonames API and return the name of a Canton if it appears in the search, otherwise it returns "no canton found"

query_city query the full text search geonames API and return the longitude and latitude of a city if there is a city in the research term, otherwise it returns "no city found"

canton_geolocalization rquery the extendedfindnearby toponym geonames API and return the canton name if it appears in the places searched, otherwise it returns "no canton found"


```python
def query_term(term):
    """query_term query the full text search the geonames API and return the name of a Canton if it appears in the search"""
    
    #Terms exluded from the search because they returns false canton.
    if term != '-' and term != 'de' and term != 'di' and term != 'und' and term != 'of':
        

        term_value = term
        
        #etree object that retrieve the xml information
        tree = etree.parse("http://api.geonames.org/search?name=%s&maxRows=200&country=CH&username=thomvett" %  (term_value))
        #crawling into each entries retrieved
        for user in tree.xpath("/geonames/geoname/fcode"):
            #test if the fcode is ADM1
            if user.text == "ADM1":
                
                #iterate to the siblings of the tag and seach for the name tag in the xml structure
                for i in user.itersiblings(preceding=True):
                    if i.tag == 'name':
                        return i.text
                    
        return 'no canton found'
    else:
        return 'no canton found'

#Here you can see it in action
print(query_term('Bern'))
```

```python
def query_city(term):
    """query_city query the full text search geonames API and return the longitude and latitude of a city if there is a city in the research term, otherwise it returns "no city found""""
    
    
    lat_lng = []
    
    #Terms exluded from the search because they returns false canton.
    if term != '-' and term != 'de' and term != 'di' and term != 'und' and term != 'of' and term != '':
        

        term_value = term
        #etree object that retrieve the xml information
        tree = etree.parse("http://api.geonames.org/search?name_equals=%s&maxRows=100&country=CH&username=thomvett" %  (term_value))
        #crawling into each entries retrieved
        for user in tree.xpath("/geonames/geoname/fcode"):
            #test if the fcode is a city see geonames fcode documentation
            if user.text == "PPL" or user.text =="PPLA":
                
                #iterate to the siblings of the tag and seach for the name tag in the xml structure
                for i in user.itersiblings(preceding=True):
                    if i.tag == 'lng':
                        lat_lng.append(i.text)
                    if i.tag == 'lat':
                        lat_lng.append(i.text)
                        
                        
                return lat_lng
                        
                    
        return 'no city found'
    else:
        return 'no city found'
    

    
#Here you can see it in action
a = query_city('Lausanne')
print(a)
print(type(a[1]))
```

```python
def canton_geolocalization(lat, lng):
    
    lat_value = lat
    lng_value = lng
    
    #etree object that retrieve the xml information
    tree = etree.parse("http://api.geonames.org/extendedFindNearby?lat=%s&lng=%s&username=thomvett" %  (lat_value, lng_value))
    #crawling into each entries retrieved
    for user in tree.xpath("/geonames/geoname/fcode"):
            
            if user.text == "ADM1":
                #iterate to the siblings of the tag and seach for the name tag in the xml structure
                for i in user.itersiblings(preceding=True):
                    if i.tag == 'name':
                        return i.text
                   
    return 'no canton found'

#Here you can see it in action. It uses the lat and lng of the test for "Lausanne" request
b = canton_geolocalization(a[1], a[0])
print(b)
```

Now we will execute the strategies explained before. It is the execution part that uses the previously defined functions.

```python
#Here you can look how the dataframe entries are splited.

print(fundingUni.index.shape)

for i in range(fundingUni.index.shape[0]):
    
    for term in fundingUni.index[i].split():
        print(term)

```

Here is the main function that execute our strategies.

```python
print(fundingUni.index.shape)
Test = []

#iterate over each dataframe entries
for i in range(fundingUni.index.shape[0]):
    
    canton_name = 'not found'
    
    #split the dataframe university name inton its terms
    for term in fundingUni.index[i].split():
        
        #query each term to see if it returns a canton
        result = query_term(term)
        if result != 'no canton found':
            canton_name = result
    #if the first strategy fail, try the second strategy
    if canton_name == 'not found':
        #query each term to see if it returns a city (the coordinates)
        for term_2 in fundingUni.index[i].split():
            coor = query_city(term_2)
            
            #if it returns coordinates, then uses canton_geolocalization to find the canton name
            if coor != 'no city found':
                canton_name = canton_geolocalization(coor[1], coor[0])
                

    fundingUni['Canton'].loc[i] = canton_name
    Test.append(canton_name)
fundingUni.head()
```

First we need to know which ID's are stored in the json file that contains the information about the different cantons.

```python
CantonsJSON = json.load(open('Data/ch-cantons.topojson.json','r'))

cantonsIdentity = np.array([['ID','Canton']])
for i in CantonsJSON['objects']['cantons']['geometries']:
    cantonsID = i['id']
    cantonsName = i['properties']['name']
    cant = np.array([[cantonsID,cantonsName]])
    cantonsIdentity = np.append(cantonsIdentity,cant,axis=0)

cantonsIdentity = pd.DataFrame(cantonsIdentity)
cantonsIdentity.columns = cantonsIdentity.loc[0]
cantonsIdentity.drop(0,inplace=True)
```

Know that we know to which ID corresponds to which cantons so then we can map to which ID.
As the cantons name do not correspon 100% to the ones that are brought back by the API we first have to change them manually, so that we can perform an adequate merge.

```python
cantonsIdentity['Canton'].loc[1] = 'Zurich'
cantonsIdentity['Canton'].loc[2] = 'Bern'
cantonsIdentity['Canton'].loc[23] = 'Valais'
cantonsIdentity['Canton'].loc[18] = 'Canton Grisons'
cantonsIdentity['Canton'].loc[3] = 'Lucerne'
cantonsIdentity['Canton'].loc[25] = 'Geneva'
cantonsIdentity['Canton'].loc[17] = 'Saint Gallen'
cantonsIdentity['Canton'].loc[11] = 'Solothurn'
```

We also have to delete the row that contains no cantons.

```python
#fundingCantons.drop('-',axis=0,inplace=True)
fundingCantons['Canton'] = fundingCantons.index
fundingCantons.head()
```

Now that this is done correctly we can merge with the university data.

```python
fundingCantons = pd.merge(fundingCantons,cantonsIdentity,on ='Canton',how='outer')
```

And we can finish by plotting the DataFrame into a chloropeth map.

```python

Swissmap = folium.Map(location=[48, -102], zoom_start=3)
Swissmap.choropleth(geo_path=r'Data/ch-cantons.topojson.json', 
                     data=fundingCantons,
                     columns=['ID', 'Approved Amount'],
                     key_on='feature.id',
                     topojson='objects.cantons',
                     fill_color='YlGn',fill_opacity=0.7, line_opacity=0.2,
                     legend_name = 'Amount of money'
                    )

Swissmap
```
